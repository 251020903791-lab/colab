```python
n = int(input("Nhấp số nguyên dương n :"))
tong =0
if n <= 0 :
  print("Vui lòng nhập số nguyên dương!")
else :

    for i in range (n+1):
      tong += i
print(f"Tổng các số từ 0 đến {n} là :{tong}")
```
