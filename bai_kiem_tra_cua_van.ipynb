```python
while True:
    try:
        n = int(input("Nhập số lượng phần tử (n) của mảng: "))
        if n <= 0:
            print("Số lượng phần tử phải là số nguyên dương (> 0). Vui lòng nhập lại.")
        else:
            break
    except ValueError:
        print("Đầu vào không hợp lệ. Vui lòng nhập một số nguyên.")
mang = []
print(f"Bắt đầu nhập {n} phần tử:")
for i in range(n):
    while True:
        try:
            # Nhập từng phần tử
            phan_tu = int(input(f"Nhập phần tử thứ {i + 1}: "))
            mang.append(phan_tu)
            break
        except ValueError:
            print("Đầu vào không hợp lệ. Vui lòng nhập một số nguyên.")
tong_chan = 0
tong_le = 0
for phan_tu in mang:
    # Kiểm tra tính chẵn lẻ bằng toán tử modulo (%)
    if phan_tu % 2 == 0:
        # Nếu là số chẵn
        tong_chan += phan_tu
    else:
        # Nếu là số lẻ
        tong_le += phan_tu
print("\n--- KẾT QUẢ ---")
print(f"Mảng đã nhập: {mang}")
print(f"Tổng các phần tử chẵn: {tong_chan}")
print(f"Tổng các phần tử lẻ: {tong_le}")
```
