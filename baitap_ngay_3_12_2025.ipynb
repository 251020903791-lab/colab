```python
a = float(input("Nhập hệ số a:"))
b = float(input("Nhập hệ số b:"))
c = float(input('Nhập hệ số c:'))

if a == 0:
  print("a = 0, phương trình trở thành hệ phương trình bậc nhất hoặc không hợp lệ.")
  if b == 0:
    if c == 0:
      print("phương trình có vô số nghiệm")
    else:
      print("phương trình vô nghiệm")
  else:
    x = -c / b
    print(f"phương trình có nghiệm duy nhất: x = {x}")
else:
  delta = b**2 - 4*a*c
  print(f"Delta = {delta}")
  if delta > 0:
    x1 = ((-b + delta**0.5) / (2*a))
    x2 = ((-b - delta**0.5) / (2*a))
    print(f"phương trình có 2 nghiệm phân biệt : x1 = {x1}, x2 = {x2} ")
  elif delta == 0:
    x = -b / (2*a)
    print(f"phương trình có nghiệm kép : x = {x}")
  else:
    # delta < 0 is implied here
    print("phương trình vô nghiệm trong tập số thực")
```
